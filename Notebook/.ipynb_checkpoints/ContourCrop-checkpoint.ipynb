# Contour Cropping

DexiNed pretrained edge detection modelinin çıktısı üzerinde contour ile croplama işlemi

```python
import numpy as np
import cv2
import imutils
import matplotlib.pyplot as plt
```

```python
def show(before, after):
    plt.figure(figsize=(16,8))
    plt.subplot(121)
    plt.imshow(before, cmap="gray")
    plt.title("Before")
    plt.subplot(122)
    plt.imshow(after, cmap="gray")
    plt.title("After")
```

```python
# Görüntülerin okunması
img = cv2.imread("../data/kimlik6.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
edge = cv2.imread("../result/BIPED2CLASSIC/avg/kimlik6.png", 0)
show(img, edge)
```

```python
# Gürültü azaltmak için blur
blur = cv2.GaussianBlur(edge, (7,7), 0)
show(edge, blur)
```

```python
# Otsu threshold
_, threshold = cv2.threshold(blur,127,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
show(blur, threshold)
```

```python
# Dilation ile kenarlar belirginleşiyor
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))
dilation = cv2.dilate(threshold, kernel, iterations=1)
show(threshold, dilation)
```

```python
# Canny Edge Detection ile kenar tespiti (Contour yakalama için gerekli)
(mu, sigma) = cv2.meanStdDev(dilation)
canny = cv2.Canny(dilation, (mu[0][0] - sigma[0][0]), (mu[0][0] + sigma[0][0]))
show(dilation, canny)
```

```python
# Dilation ile kenarlar belirginleşiyor, ufak aralıklar kapanıyor
kernel2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(4,4))
dilation2 = cv2.dilate(canny, kernel2, iterations=1)
show(canny, dilation2)
```

```python
def coords_sort(coords):
    """
    Çıktı koordinatların sıralanması için
    
    Örnek:
    [[10,100],[100,10],[100,100],[10,10]] --> [[10,10],[100,10],[10,100],[100,100]]
    """
    sum = np.sum(coords, axis=1)

    # First
    first_index = np.where(sum == min(sum))
    first = coords.pop(first_index[0][0])

    sum = np.sum(coords, axis=1)
    # Last
    last_index = np.where(sum == max(sum))
    last = coords.pop(last_index[0][0])

    # Second and Third
    if coords[0][0] > coords[1][0]:
        second = coords[0]
        third = coords[1]
    else:
        second = coords[1]
        third = coords[0]

    return first, second, third, last
```

```python
# Contour'lar yakalanıyor
cnts = cv2.findContours(dilation2, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
cnts = imutils.grab_contours(cnts)
# Contour'lar alanlarına göre sıralanıyor (Büyükten küçüğe)
cnts = sorted(cnts, key = cv2.contourArea, reverse = True)

for c in cnts:
    # Contour tamamlanması için noktalar tanımlanıyor
    peri = cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, 0.02 * peri, True)
    # Nokta sayısı 4 ise istenilen dikdörtgenimizi bulduğumuz anlamnıa geliyor
    if len(approx) == 4:
        area = cv2.contourArea(c)
        img_area = dilation2.shape[0]*dilation2.shape[1]
        # Bulunan alan ile görselin alanı arasında doğrulama
        if area > img_area/2 and area < img_area:
            # Koordinatlar sıralanıyor
            coord_list = [i[0].tolist() for i in list(approx)]
            first, second, third, last = coords_sort(coord_list)
            # Yeni ehliyet, yeni kimlik ve eski ehliyet için önerilen boyutlar (OCR modeli de daha iyi çalışıyor bu şekilde)
            maxWidth = 1540
            maxHeight = 1000
            # Perspektif fonksiyonu ile bulunan 4 nokta kırpılıp istenilen formata getiriliyor
            pts1 = np.float32([first, second, third, last])
            pts2 = np.float32([[0, 0], [maxWidth, 0], [0, maxHeight], [maxWidth, maxHeight]])
            matrix = cv2.getPerspectiveTransform(pts1, pts2)
            imgWarpColored = cv2.warpPerspective(img, matrix, (maxWidth, maxHeight))
            plt.imshow(imgWarpColored)
```
